___Load_ `tugas.csv` pada folder `data`__. Jadikan kolom `datetime` sebagai _index_

```python
### jawaban anda disini
import pandas as pd
df_tugas = pd.read_csv('data/tugas.csv', index_col='datetime')
df_tugas.head()
```

1. Dengan `map`, ubah nilai pada kolom berikut: 
 * `season`: 1 menjadi "_summer_", 2 menjadi "_autumn_", 3 menjadi "_winter_", 4 menjadi "_rainy_"
 * `holiday`: 0 untuk "_non-holiday_", 1 untuk "_holiday_"
 * `workingday`: 0 untuk "_breakday_", 1 untuk "_workingday_"
 * `weather`: 1 menjadi "_clear_", 2 menjadi "_cloudy_", 3 menjadi "_rainy_", 4 menjadi "_stormy_"

```python
### jawaban anda disini
df_tugas['season'] = df_tugas['season'].map({1:"summer",2:'autumn',3:'winter',4:'rainy'})
df_tugas['holiday'] = df_tugas['holiday'].map({0:'non-holiday',1:'holiday'})
df_tugas['workingday'] = df_tugas['workingday'].map({0:'breakday',1:'workingday'})
df_tugas['weather'] = df_tugas['weather'].map({1:'clear',2:'cloudy',3:'rainy',4:'stormy'})
df_tugas.head()
```

2. Dengan `applymap` dibantu dengan _slicing_ kolom, ubah seluruh nilai numerik dengan nilai kuadratnya menggunakan lambda

```python
### jawaban anda disini
#print (df_tugas['temp'].dtype)
df_tugas.applymap(lambda x: x**2 if type(x) == float else x)
df_tugas.head()
```

4. _Aggregate_ `data` berdasarkan `season`, `holiday`, dan `weather` untuk mendapatkan nilai maksimum dan minimum setiap kolom

```python
### jawaban anda disini
df_tugas.groupby(['season','holiday','weather']).agg([np.min,np.max])
```

5. _Aggregate_ `data` berdasarkan `season`, dan `weather` dengan fungsi berikut, dan ubah nama kolomnya:
 * _Mean_ untuk `temp` menjadi `mean_temperature`
 * _Min_ dan _max_ untuk `count` menjadi `minimal` dan `maximal`
 * Total kuadrat dari `humidity` dengan fungsi sendiri atau lambda, menjadi `humidity peak`

```python
### jawaban anda disini
df_tugasgroup = df_tugas.groupby(['season','weather']).agg({'temp': np.mean,
                                            'count': [np.min, np.max],
                                            'humidity': lambda x: np.sum(x ** 2)})
df_tugasgroup.columns = ['mean_temperature','minimal','maximal','humidity peak']
df_tugasgroup
```

6. Kelompokkan data berdasarkan semua kolom kategorik. Kemudian, untuk tiap kolom numerik, transformasikan semua data sehingga nilainya berubah menjadi rasio antara mean dan median. (Boleh membuat _function_ atau menggunakan _lambda_)

7. Budi adalah seorang analis data junior yang bekerja di sebuah perusahaan startap. Suatu hari, timnya ingin menganalisis data berupa string. Namun, pada proses analisis data tersebut karakter yang boleh muncul hanya (a-z, A-Z, dan 0-9). Budi merasa kebingungan untuk mengecek hal tersebut. Buatlah fungsi dengan menggunakan Regex untuk membantu Budi. Berikan juga defense agar jika dimasukkan integer menjadi tetap benar. 

```python
import re
def cek_karakter(string):
    kata = str(string)
    pattern = "[\W]" # Kode anda disini
    return re.search(pattern, kata) # True jika hanya mengandung karakter tsb, False jika tidak
        
print(cek_karakter("ABCDEFabcdef123450")) # Should be True
print(cek_karakter("LoremIpsumDolorSitAmet")) # Should be True
print(cek_karakter(81213177)) # Should be True
print(cek_karakter("Budi punya 5 mangga + 2 jambu")) # Should be False
```

8. Buatlah fungsi untuk mengecek apakah sebuah urutan string terdiri dari huruf kapital yang diikuti oleh huruf kecil.

```python
import re
def cek_teks(teks):
    pattern = "^[A-Z][a-z]" # Kode anda disini
    return re.search(pattern, teks) # True jika hanya mengandung karakter tsb, False jika tidak 
    
print(cek_teks("AaBbCc")) # True
print(cek_teks("Python")) # True
print(cek_teks("linux")) # False
print(cek_teks("PACMANN")) # False
```

9. Terkadang, dalam proses pengolahan data berupa teks kita hanya ingin mengambil data teks yang mengandung kata tertentu. Buatlah fungsi yang mengecek apakah suatu teks memiliki kata kunci yang dicari.

```python
# Boleh memakai regex maupun tidak
def cari_kata(kalimat,kata):
    pattern = kata # Kode anda disini
    return re.search(pattern, kalimat) # True jika hanya mengandung karakter tsb, False jika tidak 

print(cari_kata('budi membeli permen', 'permen')) # True
print(cari_kata('Jumlah pasien corona di jakarta semakin meningkat', 'surabaya')) # False
```
